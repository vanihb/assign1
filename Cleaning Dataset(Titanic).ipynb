# DATA CLEANING

# FILL AGE WITH MEAN

```python
df.Age.fillna(df.Age.mean(),inplace=True)
```

```python
df.Age
```

```python
#verify
df.Age.isna().sum()
```

# FILL AGE WITH MEDIAN

```python
df.Age.fillna(df.Age.median())
```

# Ffill and Bfill

```python
df.Fare = df.Fare.bfill()#it will fill the null values with previous data
df.Fare.isna().sum()
```

```python
df.Fare
```

```python
df.Fare = df.Fare.ffill()#it will fill the null values with forward data
df.Fare.isna().sum()
```

```python
df.Fare
```
